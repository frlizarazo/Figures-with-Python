# CÓDIGO PARA LA CREACIÓN DE UNA GRAFICA DE LINEA SIMPLE

## Librerías usadas

```python
from matplotlib.pyplot import subplots, show    # Para crear la gráfica
from os import makedirs                         # Para crear la carpeta donde se guardan las gráficas
```

## Definición de la función

```python

def create_Figure(
        Data,                          # Diccionario con los datos
        title  = 'Simple Lineplot',    # Titulo de la figura
        xlabel = 'X axis',             # Titulo del eje x
        ylabel = 'y axis',             # Titulo del eje y
        color  = 'dodgerblue',         # Color de la figura
        tickspacing = 4,               # Espaciado entre las marcas del eje horizontal
        save_title  = None,            # En caso de que el titulo sea muy largo de define un titulo
    ):                                 # secundario, para guardar la figura con ese nombre

    # Desempaquetamos los datos dados a la función
    years  = list(Data.keys())
    values = list(Data.values())

    # Se crean la figura los ejes
    fig, ax = subplots(1,1, figsize = [10,5], dpi = 1200)

    # Hacemos que la linea superior y derecha de la gráfica sean invisibles
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Hacemos que las marcas de los ejes esten "entre" los ejes y definimos un margen
    ax.tick_params(axis='both', direction='inout', pad=5)


    # Se dibuja una linea que representa los datos y puntos para especificar entre que datos se 
    # esta dibujando la linea
    ax.plot   (years, values, linewidth = 2.5, color = color, zorder = 10)
    ax.scatter(years, values, s = 20, color = color)

    # Se define el espaciamiento de las marcas
    ticks = list(range(min(years), max(years) + tickspacing + 1, tickspacing))

    ax.set_xticks(ticks)
    ax.set_xticklabels(ticks, fontsize = 8)#,rotation = 45)
    ax.set_xlim([min(ticks)-1, max(ticks)])

    # Agregamos los titulos a la figura y a los ejes
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)

    # Hacemos visible la grilla solo en el eje y
    ax.grid(axis = 'y')

    # Se intenta crear la carpeta Figuras (exist_ok es para que no haya problema si el directorio ya
    # habia sido creado anteriormente)
    makedirs('Figuras/', exist_ok = True)

    # Se guarda la figura en la carpeta Figuras con el titulo como nombre y se muestra en pantalla 
    title = save_title if save_title else title

    fig.savefig(f'Figuras/{title}.png', bbox_inches='tight', pad_inches=0.15, dpi = 1200)
    show()
```

## Ejemplo de uso

```python
# Se crea un diccionario con parejas año:valor_por_año
documents_per_year = {2024:156, 2023:180, 2022:177, 2021:158, 2020:137, 2019:158, 2018:119, 
                      2017: 89, 2016: 81, 2015: 96, 2014: 79, 2013: 81, 2012: 68, 2011: 86, 
                      2010: 63, 2009: 61, 2008: 60, 2007: 55, 2006: 62, 2005: 50, 2004: 29, 
                      2003: 40, 2002: 91, 2001: 39, 2000: 47, 1999: 34, 1998: 38, 1997: 34, 
                      1996: 19, 1995: 29, 1994: 69, 1993:  7, 1992:  8, 1991: 18, 1990: 13,
                      1989: 52, 1988: 21, 1987: 10, 1986: 19, 1985: 36, 1984:  7, 1983:  9, 
                      1982: 12, 1981: 13, 1980:  4, 1979:  6, 1978:  8, 1977:  1, 1975:  6, 
                      1974:  5, 1973:  1, 1972:  1, 1971:  1, 1968:  1, 1962:  1}

create_Figure(
    documents_per_year,
    title      = 'Documents published per year containing the term "artificial reef". Based on data from Scopus.',
    xlabel     = 'Year',
    ylabel     = 'Documents per year',
    color      = 'dodgerblue',
    save_title = 'documents_per_year'
)
```
