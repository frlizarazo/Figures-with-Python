# CÓDIGO PARA LA CREACIÓN DE VARIAS GRÁFICAS DE LINEA SIMPLE (con proyección en el último valor)

## Librerías usadas

```python
from matplotlib.pyplot import subplots, show    # Para crear la gráfica
from os import makedirs                         # Para crear la carpeta donde se guardan las gráficas
```

## Definición de la función

```python
def create_Figure(Data):
    # Desempaquetamos la información dada a la función
    labels = list(Data.keys())
    values = list(Data.values())

    Ranges = values[0]
    ranges = list(range(len(Ranges)))

    row_titles = labels[1:]
    rows       = values[1:]

    nrows = len(rows)
    ncols = len(rows[0].keys())

    # Se crean la figura y los ejes
    fig, ax = subplots(nrows,ncols, figsize = [12,8], dpi = 1200)
    fig.subplots_adjust(hspace=1.5) 

    for i, data in enumerate(rows):
        for j, (title, plot_values) in enumerate(data.items()):
            axx = ax[i, j]
            axx.spines['top'].set_visible(False)
            axx.spines['right'].set_visible(False)

            axx.plot(ranges, plot_values[:-1]               , linewidth = 2.5, zorder = 10, color = 'dodgerblue')
            axx.plot(ranges, plot_values[:-2] + [plot_values[-1]], linewidth = 2.5, zorder = 20, color = 'crimson')

            axx.scatter(ranges, plot_values[:-1]               , s = 20, zorder = 10, color = 'dodgerblue')
            axx.scatter(ranges, plot_values[:-2] + [plot_values[-1]], s = 20, zorder = 20, color = 'crimson')

            axx.grid(axis = 'y', zorder = 0)

            axx.set_xticks(ranges)
            axx.set_xticklabels(Ranges, fontsize = 8)

            axx.legend(['Datos','Datos proyectados'], loc='upper center', bbox_to_anchor=(0.5, -0.3), ncol=2)
            if j == 0:
                axx.set_ylabel('Documents published')

            if j == 1:
                title = r'$\mathbf{' + row_titles[i] + '}$\n\n' + title
            axx.set_title(title)

    # Se intenta crear la carpeta Figuras (exist_ok es para que no haya problema si el directorio ya
    # habia sido creado anteriormente)
    makedirs('Figuras/', exist_ok = True)

    # Se guarda la figura en la carpeta Figuras con el titulo como nombre y se muestra en pantalla 
    fig.savefig('Figuras/Documents per ranges.png', bbox_inches='tight', pad_inches=0.15, dpi = 1200)
    show()
```

## Ejemplo de uso
